```python
from numpy.random import uniform
from matplotlib import pyplot as plt
from numpy import log, mean, median, linspace, exp
```

```python
N=1000000
x = uniform(low=0.0, high=10, size=N)
plt.hist(x, density=True)
plt.show()
y=log(x)
plt.hist(y, bins=100, density=True)
x_plot=linspace(-2.3, 2.3, num=1000)
plt.plot(x_plot, exp(x_plot)/9.9)
plt.show()

print(f'Log of mean of x: {log(mean(x))}')
print(f'Mean of log of x: {mean(y)}')
print(f'Log of median of x: {log(median(x))}')
print(f'Mean of log of x: {median(y)}')
```
