```python
from scipy.stats import norm
from numpy import sum, sqrt, pi, abs, std
from matplotlib import pyplot as plt
```

```python
integral=[]

for N in range(1,30000):
    gaussian = norm(loc=0 , scale=1)
    x = gaussian.rvs(N)
    integral.append(sum(abs(x)**3)/(2*N)) 
```

```python
plt.plot(integral)
plt.axhline(y=sqrt(2/pi), c='r')
plt.show()

plt.plot(abs(integral-sqrt(2/pi)))
plt.show()
```

```python
integral1=[]

N=1000

for i in range(10000):
    gaussian = norm(loc=0 , scale=1)
    x = gaussian.rvs(N)
    integral1.append(sum(abs(x)**3)/(2*N)) 

N=10000
integral2=[]

for i in range(10000):
    gaussian = norm(loc=0 , scale=1)
    x = gaussian.rvs(N)
    integral2.append(sum(abs(x)**3)/(2*N)) 
```

```python
plt.hist(integral1, bins=100, color='r')
plt.show()
plt.hist(integral2, bins=100, color='b')
plt.show()

print(std(integral1))
print(std(integral2))
```
